```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", color_codes=True)
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Read Dataset and save it in a Pandas Dataframe
dataset = pd.read_csv("Iris.csv")

# Taking a look at first few rows
dataset.head(10)
```

```python
# Dimensions of the input data(number of rows and columns)
print("Iris dataframe shape is :"+str(dataset.shape))
```

```python
dataset.describe()
```

```python
# Check unique values in Species Column
dataset["Species"].value_counts()
```

```python
# Compare "SepalLengthCm" and "SepalWidthCm" through scatterplot
dataset.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
ax = sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=dataset)
```

```python
# Comparing "SepalLengthCm" to "Species" through sns kernel density plot
ax = sns.FacetGrid(dataset, hue="Species", size=5).map(sns.kdeplot, "SepalLengthCm").add_legend()
```

```python
# Compare both "SepalLengthCm" and "SepalWidthCm" to "Species" column through Scatter plot
ax = sns.FacetGrid(dataset, hue="Species", size=5).map(plt.scatter, "SepalLengthCm", "PetalLengthCm").add_legend()
```

```python
# Compare relation thorugh Violin plot
ax = sns.violinplot(x="Species", y="PetalLengthCm", data=dataset)
```

```python
# Compare relation through seaborn pairplot with kde as diagonal plots
ax = sns.pairplot(dataset.drop("Id", axis=1), hue="Species", diag_kind="kde")
```

```python
# Exploring relation between features through pandas multivariate visualisation techniques
from pandas.tools.plotting import andrews_curves
andrews_curves(dataset.drop("Id", axis=1), "Species")
```

```python
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(dataset.drop("Id", axis=1), "Species")
```

```python
# 2D plane graph
from pandas.tools.plotting import radviz
radviz(dataset.drop("Id", axis=1), "Species")
```
